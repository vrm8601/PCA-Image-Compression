```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  
from numpy import linalg as LA
from sklearn.preprocessing import scale
%matplotlib inline
from sklearn.model_selection import train_test_split 
from sklearn.datasets import fetch_openml
X,y=fetch_openml('mnist_784', version=1, return_X_y=True)
np.random.seed(1)
X_train, X_test, y_train, y_test = train_test_split(X, y)
X_train=scale(X_train)
X_test=scale(X_test)
```

# Find a SVD for the data.

__Part A__ What is the  shape of $X_{train}$?

```python
X_train.shape
```

Now Lets find the SVD of $M=X_{train}^TX_{train}$  <font color=red> Note: We could take the SVD of X_train but it is painfully slow.</font> Use np.dot or A.dot(B) for all matrix multiplications.

```python
M=np.dot(X_train.T,X_train)
U,s,VT=np.linalg.svd(M)
S=np.diag(s)
```

__Part B__  Check the norm of $M-USV^T$.  It should be close to zero

```python
LA.norm(M-U.dot(S.dot(VT)))
```

__Part C__ Lets plot the singular values.   The sum of all of them is the variance so let's sum them as well.

```python
plt.plot(s)
np.sum(s)
```

Our goal is to keep 90% of the total variance. The code below plots a cumulative sum of the variance and shows that if we keep the first 234 singular values we have 90% of the variance.       

```python
DS=np.cumsum(s)/np.sum(s)
V=.9
d=np.argmax(DS>=V)+1
plt.plot(DS)    
h_line = .98*np.ones([728])
plt.plot(h_line)
plt.axvline(x=d,color='r')
print("d=",d)
```

# Using the SVD to reduce the data- Projection 


From 3 dimensions to two dimensions



to 2 dimensions



How do we do this? We simply multiply our data by a sub matrix of U.    

__Part D__ Multiply $X_{train}$ by $U_d$ which corresponds to the first 154 columns of $U$.  How much have you reduced the data? Lets call it $X_{reduced}=X_{train}U_d$ 

```python

U_d=U[:,:d] # remember order kept the indecies of the largest eigenvalues
XU_d=np.dot(X_train,U_d)
```

```python
XU_d.shape
```

```python
d/784
```

To recover we simply need to multiply again on the right by $U_d^T$ that is $X_{recovered}=(X_{train}U_d)U_d^T.$ 

__Part E__ Do the dimensions come out  right?  Compute the difference between the matrices like below  

$$\frac{\|X_{train}-X_{recovered}\|_2^2}{\|X_{train}\|_2^2}$$




```python
err=np.linalg.norm(X_train-np.dot(XU_d,U_d.T))**2/np.linalg.norm(X_train)**2
print(err)
```

# SKlearn PCA

```python
from sklearn.decomposition  import PCA
pca=PCA()
```

```python
pca.fit(X_train)
csum=np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(csum>=V)+1
pca=PCA(n_components=d)
X_train_reduced=pca.fit_transform(X_train)
```

```python
sgd_clf = SGDClassifier(random_state=42, max_iter=100,tol=-np.infty)
sgd_clf.fit(X_train_reduced, y_train)
```

```python
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(sgd_clf, X_train_reduced, y_train, cv=3, scoring="accuracy")
print(cvs)
```
